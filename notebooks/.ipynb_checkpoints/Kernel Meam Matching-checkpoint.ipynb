```python
# from https://github.com/vodp/py-kmm

import os
import sys
import numpy as np 
import math
import matplotlib.pyplot as plt
from cvxopt import matrix, solvers
from scipy import sparse

%matplotlib inline
```

```python
# an implementation of Kernel Mean Matchin
# referenres:
#  1. Gretton, Arthur, et al. "Covariate shift by kernel mean matching." Dataset shift in machine learning 3.4 (2009): 5.
#  2. Huang, Jiayuan, et al. "Correcting sample selection bias by unlabeled data." Advances in neural information processing systems. 2006.
def kernel_mean_matching(X_train, X_test, kern='lin', B=1.0, eps=None):
    nx = X_train.shape[0]
    nz = X_test.shape[0]
    
    print(nx, nz)
    if eps == None:
        eps = B/math.sqrt(nz)
    if kern == 'lin':
        K = sparse.csr_matrix.dot(X_test, X_test.T)
        kappa = np.sum(sparse.csr_matrix.dot(X_test, X_train.T)*float(nz)/float(nx),axis=1)
    elif kern == 'rbf':
        K = compute_rbf(X_test,X_test)
        kappa = np.sum(compute_rbf(X_test,X_train),axis=1)*float(nz)/float(nx)
    else:
        raise ValueError('unknown kernel')
        
    print("K ", K.shape, type(K))
    K = matrix(K)
    print("kappa ", kappa.shape)
    kappa = matrix(kappa)
    
    G = matrix(np.r_[np.ones((1,nz)), -np.ones((1,nz)), np.eye(nz), -np.eye(nz)])
    h = matrix(np.r_[nz*(1+eps), nz*(eps-1), B*np.ones((nz,)), np.zeros((nz,))])
    
    sol = solvers.qp(K, -kappa, G, h)
    coef = np.array(sol['x'])
    return coef

def compute_rbf(X_train, X_test, sigma=1.0):
    K = np.zeros((X_train.shape[0], X_test.shape[0]), dtype=float)
    print("*", K.shape)
    for i, vx in enumerate(X_train):
        K[i,:] = np.exp(-np.sum((vx-X_test)**2, axis=1)/(2.0*sigma))
    return K
```

```python
x = 11*np.random.random(200)- 6.0
y = x**2 + 10*np.random.random(200) - 5
X_test = np.c_[x, y]

print(x.shape, y.shape)
print("X_test: ", X_test.shape)

x = 2*np.random.random(10) - 6.0
y = x**2 + 10*np.random.random(10) - 5
X_train = np.c_[x, y]

print()
print(x.shape, y.shape)
print("X_train: ", X_train.shape)
print()

coef = kernel_mean_matching(X_train, X_test, kern='rbf', B=10)

#plt.close()
#plt.figure()
#plt.scatter(Z[:,0], Z[:,1], color='black', marker='x')
#plt.scatter(X[:,0], X[:,1], color='red')
#plt.scatter(Z[:,0], Z[:,1], color='green', s=coef*10, alpha=0.5)

np.sum(coef > 1e-2)
```

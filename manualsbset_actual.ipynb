```python
def clean(timelines):
    i = 0
    l = 0
    for t in timelines:
        if '"urls":' in t or '"media":' in t or '[{"url":' in t or "RT" in t:
            del timelines[i]
            l += 1
        i += 1
        
    return l
```

```python

def write_timeline(timeline, filename, label):
    fname = '../extended_manual/'+filename+'_extended.txt'
    
    with open(fname, 'a+') as f:
        f.write(label)
        
        numtweets = 0
        for tweet in timeline:
            for word in tweet:
                #print word
                f.write(word)
                f.write(' ')
            numtweets += 1
            if numtweets == 200:
                f.write('\n')
                f.write(label)
                numtweets = 0
                
            f.write('\t')
```

```python
def create_list_offilestowrite(file):
    filenames = []
    csv_data = pd.read_csv(file,low_memory = False)
    df = pd.DataFrame(csv_data)
    group1 = df.groupby(['country']).count()
    countries = df['country'].unique()
    
    for country in countries:
        country = country.strip('\'')
        #print(country)
        filename = '../locations_manual/countries/manual_country_'+country+'.txt'
        filenames.append(filename)

    return filenames
    #print(filenames)
```

```python
def create_list_offilestowrite_states(file):
    filenames = []

    csv_data = pd.read_csv(file,low_memory = False)
    df = pd.DataFrame(csv_data)
    #group1 = df.groupby(['state']).count()
    states = df['state'].unique()
    
    for state in states:
        state = state.strip('\'')
        state = state.strip('\'')
        #print(country)
        filename = '../locations_manual/states/manual_state_'+state+'.txt'
        filename = filename.replace("'", "")
        filename = filename.replace(" ", "")
        
        filenames.append(filename)

    return filenames
    #print(filenames)
```

```python
def subset_ids_bycountry(file):
    ids_bycountry = []

    csv_data = pd.read_csv(file,low_memory = False)
    df = pd.DataFrame(csv_data)
    group1 = df.groupby(['country']).count()
    countries = df['country'].unique()
    
    countryids = []
    for country in countries:
        ids = []
        countrydf = df[df['country'] == country]
        print(countrydf)
        
        ids = countrydf['id']
        countryids.append(ids)

    return countryids
```

```python
def subset_ids_bystate(file):

    csv_data = pd.read_csv(file,low_memory = False)
    df = pd.DataFrame(csv_data)
    states = df['state'].unique()
    
    stateids = []
    for state in states:
        ids = []
        statedf = df[df['state'] == state]
        #print(countrydf)
        
        ids = statedf['id']
        stateids.append(ids)

    return stateids
```

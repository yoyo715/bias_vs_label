```python
import pandas as pd
import json
import csv
```

```python
corpus_dir='./corpora/query-gender.json'
corpus = json.load(open(corpus_dir, 'r'))
```

```python
userd = {}
ids = []
labels = []
for idx, info in corpus['annotations'].items():
    userd[idx] = info['query_label2']
    #print(info['query_label2'], idx)
    labels.append(info['query_label2'].encode('ascii','ignore'))
    ids.append(idx.encode('ascii','ignore'))
    
    
print len(ids)
```

```python
correct_labels = []

# 1 is female, 0 is male
for label in labels:
    if label == "m":
        correct_labels.append("__label__0 ")
    elif label == "f":
        correct_labels.append("__label__1 ")
    else:
        #print label
        correct_labels.append("__label__-999")
```

```python
#print labels

df = pd.DataFrame(
    {'label': labels,
     'ids': ids,
     'written_labels': correct_labels,
    })

print len(df)

new_df = df[df.label != '0']
new_df.reset_index(inplace=True) 

print new_df
print(len(new_df))
```

```python
import tweepy
from tweepy import OAuthHandler
from ACCESS_TOKENS_DONTSHARE import *   # To get twitter api authentication creds

        
CONSUMER_KEY = get_consumer_key()
CONSUMER_SECRET = get_consumer_secret()
ACCESS_KEY = get_access_token()
ACCESS_SECRET = get_access_secret()

auth = OAuthHandler(CONSUMER_KEY,CONSUMER_SECRET)
api = tweepy.API(auth)
auth.set_access_token(ACCESS_KEY, ACCESS_SECRET)

#search
api = tweepy.API(auth)
```

```python
new_df.loc[0 , "written_labels"]

users = new_df.loc[:, "ids"]

for user in users:
    print user
```

```python
# filter out retweets or no??

not_avail = 0
found = 0
l = 0
q = 0

for user in users:
    i = 0
    tweets = []
    label = new_df.loc[l , "written_labels"]
    l += 1
    
    try:
        # collect tweets from user
        for status in tweepy.Cursor(api.user_timeline, user_id=user).items():
            t = status._json['text']
            tweets.append(t)
            i += 1
            
            if i == 200:
                break
                
        found += 1
        # write tweets to file
        if i < 200:
            print "User has less than 200 tweets"
            q += 1
        else:   
            with open('../self_labeled_set_COLLECTED.txt', 'a') as f:
                f.write(label)
                for item in tweets:
                    item = item.encode('ascii','ignore').replace("\n", "")
                    f.write("%s\t" % item)
                f.write("\n")  
    except Exception:
        not_avail += 1
        pass
    
    
print not_avail, found, q
```
